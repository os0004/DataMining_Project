```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Loading Dataset
df = pd.read_csv("movies.csv")
```

```python
df.drop_duplicates(subset="MOVIES", keep=False, inplace=True)
```

```python
df.shape
```

```python
df['GENRE'] = df['GENRE'].str.replace("n", " ").str.strip()
df['ONE-LINE'] = df['ONE-LINE'].str.replace("n", " ").str.strip()
df['STARS'] = df['STARS'].str.replace("n", " ").str.strip()
```

```python
df
```

```python
df['Year'] = df['YEAR'].str.extract(r'([0-9]{4}–.*|[0-9]{4})')
df['Year'] = df['Year'].str.strip().replace(")", "")
```

```python
# Extracts year from "- )" cases

def extract_year(year):
    if pd.notna(year) and year.endswith('– )'):
        return year.replace('– )', '–')
    else:
        return year

# Apply the 'extract_year' function to the 'Year' column and fill missing values with 'Unknown'
df['Year'] = df['Year'].apply(extract_year).fillna('Unknown')
```

```python
df
```

```python
# Fill missing values in GENRE with Unknown
df['GENRE'] = df['GENRE'].fillna('Unknown')
```

```python
# Filling missing values in 'RATING' column with the mean of that column
df['RATING'].fillna(int(df['RATING'].mean()), inplace=True)
```

```python
# Replacing ',' with '' in column VOTES
df['VOTES'] = df['VOTES'].str.replace(',', '')

# Changing the datatype of column VOTES from object to float
df['VOTES'] = pd.to_numeric(df['VOTES'])

# Filling missing values in VOTES column with mean values of the column
df['VOTES'].fillna((df['VOTES'].mean()), inplace=True)
```

```python
# datatype of VOTES changed to float
df.dtypes
```

```python
df
```

```python
# Filling missing values in RunTime column with mean values of the column
df['RunTime'].fillna(int(df['RunTime'].mean()), inplace=True)
```

```python
# Dropping 'Gross' and 'YEAR' columns from df
df.drop(columns=['Gross', 'YEAR'], inplace=True)
```

```python
def extract_director(direc):
    if 'Director' in direc or 'Directors' in direc:
        director = direc.strip().split("|")[0]  # The Second Half is the stars
        return director.split(":")[1].strip()  # Return the Director name
    else:
        return ''

def extract_stars(stars):
    if 'Star' not in stars or 'Stars' not in stars:
        return ''
    else:
        return stars.split(":")[-1].strip()  # last value in this list will be the stars

df['Director'] = df['STARS'].apply(lambda d: extract_director(d))
df['Stars'] = df['STARS'].apply(lambda s: extract_stars(s))
```

```python
df['Director'].fillna('Unknown', inplace=True)
df['Stars'].fillna('Unknown', inplace=True)
```

```python
from scipy import stats

# Handling Outliers for 'RATING' column
z_scores_rating = np.abs(stats.zscore(df['RATING']))
threshold = 3
outliers_rating = df[z_scores_rating > threshold]
df_cleaned = df[z_scores_rating <= threshold]

num_outliers_rating = len(df) - len(df_cleaned)
print(f"Number of outliers removed from 'RATING': {num_outliers_rating}")

# Handling Outliers for 'VOTES' column
z_scores_votes = np.abs(stats.zscore(df['VOTES']))
threshold = 3
outliers_votes = df[z_scores_votes > threshold]
df_cleaned = df_cleaned[z_scores_votes <= threshold]

num_outliers_votes = len(df) - len(df_cleaned)
print(f"Number of outliers removed from 'VOTES': {num_outliers_votes}")

# Validate the changes made
# Check for outliers in 'RATING' and 'VOTES' after removal
z_scores_rating_cleaned = np.abs(stats.zscore(df_cleaned['RATING']))
z_scores_votes_cleaned = np.abs(stats.zscore(df_cleaned['VOTES']))

print(f"Outliers in 'RATING' after removal: {z_scores_rating_cleaned[z_scores_rating_cleaned > threshold]}")
print(f"Outliers in 'VOTES' after removal: {z_scores_votes_cleaned[z_scores_votes_cleaned > threshold]}")
```
